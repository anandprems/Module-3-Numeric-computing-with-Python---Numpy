```python
#loading image
import numpy as np
from PIL import Image

# Load image
image = Image.open('apj.png')

# Convert image to NumPy array
image_array = np.array(image)

print(image_array)
```

```python
#displaying image
import matplotlib.pyplot as plt

# Display image
plt.imshow(image_array)
plt.axis('off')
plt.show()
```

```python
#Resizing an Image:
import cv2
# Resize image
new_width = 500
new_height = 200
resized_image = cv2.resize(image_array, (new_width, new_height))

# Display image
plt.imshow(resized_image)
plt.axis('off')
plt.show()
```

```python
# Crop image
x = 100  # starting x-coordinate
y = 50   # starting y-coordinate
width = 500
height = 150
cropped_image = image_array[y:y+height, x:x+width]

# Display image
plt.imshow(cropped_image)
plt.axis('off')
plt.show()
```

```python
#applying filter
import scipy.ndimage

# Apply Gaussian filter
sigma = 1.5
filtered_image = scipy.ndimage.gaussian_filter(image_array, sigma=sigma)

# Display image
plt.imshow(filtered_image)
plt.axis('off')
plt.show()
```

```python
# Convert image to grayscale
grayscale_image = cv2.cvtColor(image_array, cv2.COLOR_RGB2GRAY)

# Display image
plt.imshow(grayscale_image)
plt.axis('off')
plt.show()
```

```python
# Save image
output_image = Image.fromarray(filtered_image)
output_image.save('filtered_image.jpg')
```
